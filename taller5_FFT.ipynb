## Taller 5 - Transformada rápida de Fourier (FFT)
#### Implementación de la FFT de numpy 

```python
import numpy as np 
import numpy.random as npr
import matplotlib.pyplot as plt
plt.style.use('classic')
#plt.style.use('seaborn-poster')
```

```python
class Wave: 
    def __init__(self): 
        # Amplitud, Desfase, freq Angular
        self.params  = [npr.rand(),npr.rand(),npr.rand()]
    def evaluate(self,x):
        return self.params[0]*np.sin(self.params[1]+2*np.pi*x*self.params[2]) 
```

```python
def main():
    n_waves = 30 
    waves = [Wave() for  i in range(n_waves)]
    x = np.linspace(-10,10,500)
    y = np.zeros_like(x)
    
    for wave in waves:
        y += wave.evaluate(x)
     
   # plt.plot(y)
    #plt.show()
    
    #Transformada rápida de fourier de numpy
    f=np.fft.fft(y)
    freq = np.fft.fftfreq(len(y),d=x[1]-x[0])
        
    fig, ax = plt.subplots(2)
    for wave in waves: 
        ax[0].plot(wave.evaluate(x), color='black',alpha=0.3)
        ax[0].plot(y,color='blue')
        ax[1].plot(freq,abs(f)**2)
        
        plt.show
    
if __name__ == '__main__':
    main()
```
